```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Create random numbers
x = np.linspace(0,10)

# Create empty canvas
fig = plt.figure()

# Add axes to figure
axes = fig.add_axes([0,0,1,1]) # left, bottom, width, height (range 0 to 1)
axes.plot(x,x**2,label = "x**2")

# Add labels & titles
axes.set_xlabel('X label here')
axes.set_ylabel('Y label here')
axes.set_title('Title here')

# Legend
axes.legend(loc=0,frameon=False)
```

```python
# Change canvas size, resolution and frame
fig = plt.figure(figsize=(8,5), dpi=200,frameon = True)
axes = fig.add_axes([0,0,1,1])

'''
Quick color bank
b	blue
g	green
r	red
c	cyan
m	magenta
y	yellow
k	black
w	white
'''

colors = ['r','#FF8C00','y','g','c','b','m','k']
 
for color in colors:
    gap = colors.index(color)
    label = 'x**2 - 10 * {}'.format(gap)
    axes.plot(x, x**2 - 10 * gap, color = color, linewidth = 5, linestyle='-',label = label)
axes.legend(loc=0,frameon = False)
axes.set_title('Somewhere over the rainbow',fontsize=24, fontweight='bold')
plt.savefig("Somewhere over the rainbow.png", bbox_inches="tight", facecolor="#ffffff")
```

```python
fig = plt.figure(figsize=(8,5), dpi=200,frameon = True)
axes = fig.add_axes([0,0,1,1])

'''
Quick line style bank
'-'	solid line style
'--'	dashed line style
'-.'	dash-dot line style
':'	dotted line style
'''

line_styles = ['-','--','-.',':']

for style in line_styles:
    gap = line_styles.index(style)
    label = 'x**2 - 10 * {}'.format(gap)
    axes.plot(x, x**2 - 10 * gap, color = 'g', linewidth = 3, linestyle= style, label = label)
axes.legend(loc=0,frameon = False)
axes.set_title('LineStyles',fontsize=16, fontweight='bold')


```

```python
fig = plt.figure(figsize=(8,5), dpi=200,frameon = True)
axes = fig.add_axes([0,0,1,1])
'''
Quick marker style bank
'.'	point marker
','	pixel marker
'o'	circle marker
'v'	triangle_down marker
'^'	triangle_up marker
'<'	triangle_left marker
'>'	triangle_right marker
'1'	tri_down marker
'2'	tri_up marker
'3'	tri_left marker
'4'	tri_right marker
's'	square marker
'p'	pentagon marker
'*'	star marker
'h'	hexagon1 marker
'H'	hexagon2 marker
'+'	plus marker
'x'	x marker
'D'	diamond marker
'd'	thin_diamond marker
'|'	vline marker
'_'	hline marker
'''
marker_styles = ['.',',','o','v','^','<','>','1','2','3','4','s','p','*','h','H','+','x','D','d','|','_']
for style in marker_styles:
    gap = marker_styles.index(style)
    label = 'x**2 - 10 * {}'.format(gap)
    axes.plot(x, x**2 - 10 * gap, color = 'g', linewidth = 1, linestyle= '-', marker = style)

axes.set_title('Marker styles',fontsize=16, fontweight='bold')
```

```python
fig = plt.figure(figsize=(8,5), dpi=200,frameon = True)
axes = fig.add_axes([0,0,1,1])

for size in range(1,10):
    #gap = marker_styles.index(style)
    label = 'x**2 - 10 * {}'.format(size)
    axes.plot(x, x**2 - 10 * size, color = 'g', linewidth = 1, linestyle= '-', marker = 'o', markersize = size, label = label)

axes.set_title('Marker sizes',fontsize=16, fontweight='bold')
axes.legend(loc = 0)
```

```python
fig = plt.figure(figsize=(8,5), dpi=200,frameon = True)
axes = fig.add_axes([0,0,1,1])
colors = ['r','#FF8C00','y','g','c','b','m','k']
for marker_face_color in colors:
    gap = colors.index(marker_face_color)
    label = 'x**2 - 10 * {}'.format(marker_face_color)
    axes.plot(x, x**2 - 10 * gap, color = 'g',
              linewidth = 1,
              linestyle= '-',
              marker = 'o',
              markersize = 10,
              markerfacecolor = marker_face_color,
              label = label)

axes.set_title('Marker face color',fontsize=16, fontweight='bold')
axes.legend(loc = 0)
```

```python
# Subplots, Log scale, Scientific notation, Show grid
fig, axes = plt.subplots(1, 2, figsize=(14,7))
      
axes[0].plot(x, np.exp(x),label="exp(x)")
axes[0].plot(x, x,label="x")
axes[0].set_title("Normal scale")
axes[0].legend(loc=0)

axes[1].plot(x, np.exp(x),label="exp(x)")
axes[1].plot(x, x,label="x")
axes[1].legend(loc=0)
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

```python
# Scatter plot with color
n = 500
X = np.random.randn(n)
Y = np.random.randn(n)
C = np.sin(Y-X)

fig = plt.figure(figsize=(8,5), frameon = False)
axes = fig.add_axes([0,0,1,1])

axes = plt.scatter(X,Y,c=C,alpha=0.3)
plt.show()
```

```python
# Histogram

mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(500)

num_bins = 50

fig, axes = plt.subplots(figsize=(10,5))

# the histogram of the data
n, bins, patches = axes.hist(x, num_bins, density=1)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
axes.plot(bins, y, '--')
axes.set_xlabel('Smarts')
axes.set_ylabel('Probability density')
axes.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```
