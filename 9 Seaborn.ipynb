```python
import seaborn as sns
%matplotlib inline
```

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
mpg.shape
```

```python
sns.distplot(mpg['mpg'])
```

```python
sns.distplot(mpg['mpg'],kde=False,bins=100)
```

```python
sns.jointplot(x='mpg',y='weight',data=mpg,kind='scatter')
```

```python
sns.jointplot(x='mpg',y='weight',data=mpg,kind='hex')
```

```python
sns.jointplot(x='mpg',y='weight',data=mpg,kind='reg')
```

```python
sns.pairplot(mpg)
```

```python
sns.pairplot(mpg,hue='cylinders',palette='coolwarm')
```

```python
sns.barplot(x='origin',y='mpg',data=mpg)
```

```python
sns.countplot(x='origin',data=mpg)
```

```python
sns.boxplot(x="origin", y="mpg", data=mpg,palette='rainbow')
```

```python
sns.stripplot(x="origin", y="mpg", data=mpg)
```

```python
sns.swarmplot(x="origin", y="mpg", data=mpg)
```

```python
sns.swarmplot(x="origin", y="mpg",data=mpg, hue='cylinders',palette="coolwarm", split=True)
```

```python
sns.boxplot(x="origin", y="mpg", data=mpg)
sns.swarmplot(x="origin", y="mpg", data=mpg, color=".25", alpha = 0.3)
```

```python
sns.violinplot(x="origin", y="mpg", data=mpg,palette='rainbow')
```

```python
sns.violinplot(x="origin", y="mpg", data=mpg,hue='cylinders',palette='coolwarm')
```

```python
sns.violinplot(x="origin", y="mpg", data=mpg,palette='rainbow')
sns.swarmplot(x="origin", y="mpg", data=mpg,color='black',size=3,alpha = 0.3)
```

```python
mpg.corr()
```

```python
sns.heatmap(mpg.corr())
```

```python
sns.heatmap(mpg.corr(),cmap='coolwarm',annot=True)
```

```python
mpg.pivot_table(values='mpg',index='model_year',columns='cylinders')
```

```python
pvmpg = mpg.pivot_table(values='mpg',index='model_year',columns='cylinders')
sns.heatmap(pvmpg)
```

```python
sns.heatmap(pvmpg,cmap='magma',linecolor='white',linewidths=1)
```

```python
sns.lmplot(x='mpg',y='weight',data=mpg)
```

```python
sns.lmplot(x='mpg',y='weight',data=mpg,hue='origin')
```

```python
sns.lmplot(x='mpg',y='weight',data=mpg,hue='origin',palette='Set1',
           markers=['.','v','+'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='mpg',y='weight',data=mpg,col='origin')
```

```python
sns.lmplot(x='mpg',y='weight',data=mpg,col='origin',hue='cylinders',palette='Set1')
```
